```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/drive/MyDrive/10k/listings.csv")
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop(['neighbourhood_group','license'], axis = 1,inplace = True)
```

```python
avg = df.groupby('neighbourhood')['price'].mean()

def fill(df):

    if pd.isnull(df['price']):

        if df['neighbourhood'] in avg:
            return avg[df['neighbourhood']]

    return df['price']

df['price'] = df.apply(fill, axis=1)
```

```python
df.dtypes
```

```python
df[df['last_review'].isnull()]
```

```python

df['reviews_per_month'].fillna(0, inplace=True)

df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce').fillna(pd.Timestamp('2024-01-01'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

palette = sns.color_palette("bright")

plt.figure(figsize=(12, 6))
sns.countplot(x='neighbourhood', data=df, palette=palette)
plt.title('Number of Listings by Neighbourhood')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(8, 6))
sns.kdeplot(df['price'], shade=True)
plt.title('Density Plot of Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()

numeric_df = df.select_dtypes(include=['int64', 'float64'])

corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price Distribution by Room Type')
plt.show()


plt.figure(figsize=(8, 6))
df['room_type'].value_counts().plot(kind='bar', color=palette)
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.show()
```

```python

cols = df.select_dtypes(include=['int64', 'float64']).columns

sc = StandardScaler()

dfN = df.copy()
dfN[cols] = sc.fit_transform(dfN[cols])
dfN
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=dfN)
plt.title('Price Distribution by Room Type')
plt.show()
```
